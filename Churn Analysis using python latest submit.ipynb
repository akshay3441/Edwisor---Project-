```python
## So there is no missing value in train and test data structure.. Now we will proceed further for other pre-processing techniques
## Now we will convert categorical and strings data into numeric data as a part of pre - processing technique for both train and test data..

lis = []
for i in range(0, data_train.shape[1]):
    #print(i)
    if(data_train.iloc[:,i].dtypes == 'object'):
        data_train.iloc[:,i] = pd.Categorical(data_train.iloc[:,i])
        #print(data_train[[i]])
        data_train.iloc[:,i] = data_train.iloc[:,i].cat.codes 
        data_train.iloc[:,i] = data_train.iloc[:,i].astype('object')
        
        lis.append(data_train.columns[i])
```

```python
## So we have sorted out the numeric data from train and test data in order to perform outlier analysis..
# Now we will perform outlier analysis and remove outlier from train and test data in order to filter out the data..

# #Detect and delete outliers from data
for i in cnames:
    print(i)
    q75, q25 = np.percentile(data_train.loc[:,i], [75 ,25])
    iqr = q75 - q25

    min = q25 - (iqr*1.5)
    max = q75 + (iqr*1.5)
    print(min)
    print(max)
  
    data_train = data_train.drop(data_train[data_train.loc[:,i] < min].index)
    data_train = data_train.drop(data_train[data_train.loc[:,i] > max].index)
```

```python
#replace target categories with Yes or No for test data
data_test['Churn'] = data_test['Churn'].replace(0, 'No')
data_test['Churn'] = data_test['Churn'].replace(1, 'Yes')
```

```python
## X1 represents independent variable for training data
## X2 represents independent variable for testing data
## Y1 represents dependent variable for training data
## Y2 represents dependent variable for testing data

X1 = data_train.values[:, 0:15]
Y1 = data_train.values[:,15]
X2= data_test.values[:, 0:15]
Y2 = data_test.values[:,15]
```
